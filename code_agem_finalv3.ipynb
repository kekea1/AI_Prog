```python
import os
import torch
import pandas as pd
from torch.utils.data import Dataset
from PIL import Image

class IntelDataset(Dataset):
    def __init__(self, df, transform=None, base_dir="/content/my_intel_dataset/seg_train/seg_train", label_map=None):
        self.df = df.reset_index(drop=True)
        self.transform = transform
        self.base_dir = base_dir
        self.label_map = label_map # Dictionnaire de conversion, ex: {3: 1, 1: 0}

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_name = row['image_name']
        label = int(row['label'])

        # On récupère le nom de la classe AVANT de remapper le label pour trouver le bon dossier
        class_name = label_to_classname[label]

        # Remappage pour que le label soit 0 ou 1 pour le modèle binaire
        if self.label_map is not None:
            label = self.label_map[label]

        img_path = os.path.join(self.base_dir, class_name, img_name)
        image = Image.open(img_path).convert('RGB')

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)
```

```python
class EpisodicMemory:
  def __init__(self, max_memories_sample=1000):
    """
    max_memory_sample : c'set le max d'image a retenir pour chaque tache passée
    """
    self.max_memories_sample=max_memories_sample
    self.memory_dataset=None   # Pas encore de TensorDataset à acceuillir

  def store_memories(self,task_dataset):
    """
    extrait un echantullon aleatoire de la tache qui veint de se terminer
    """

    # On cree un datalaoder temporaire avec comme batchsize le nombre de "souvenir" max autorisé
    temp_loader=DataLoader(task_dataset, batch_size=self.max_memories_sample,shuffle=True)
    #extraction d'un batch
# iter => actuve l'iterateur et next +> prend la "1ere boucle"
    images_from_samples,labels_sample = next(iter(temp_loader))

    if self.memory_dataset is None:
      #TensorDataste => permet de reunir les X et les y en un daatset
       #EN 0 => les x et en 1 => le y
      self.memory_dataset=TensorDataset(images_from_samples,labels_sample)

    else:
      old_labels=self.memory_dataset.tensors[1]
      old_images=self.memory_dataset.tensors[0]

      # On fusionne le passé et le present

      new_images=torch.cat([old_images,images_from_samples] )
      new_labels=torch.cat([old_labels,labels_sample])

      self.memory_dataset=TensorDataset(new_images,new_labels)

      print(f"-> Buffer A-GEM mis à jour. Taille totale de la mémoire : {len(self.memory_dataset)} images.")



  def get_reference_batch(self, batch_size):
    if self.memory_dataset is None:
      return None

    ref_loader=DataLoader(self.memory_dataset, batch_size=batch_size,shuffle=True)
    return next(iter(ref_loader))


```

```python
def train_agrem_step2(model, optimizer, criterion, current_batch, batch_ref, device):
    current_imgs, current_labels = current_batch[0].to(device), current_batch[1].to(device)
    ref_imgs, ref_labels = batch_ref[0].to(device), batch_ref[1].to(device)

    # 1. Calcul du gradient sur la tâche courante (g)
    optimizer.zero_grad()
    current_output = model(current_imgs)
    current_loss = criterion(current_output, current_labels)
    current_loss.backward()

    params = [p for p in model.parameters() if p.requires_grad]
    g = parameters_to_vector([p.grad for p in params]).clone()

    # 2. Calcul du gradient sur la mémoire (g_ref)
    optimizer.zero_grad()
    ref_output = model(ref_imgs)
    ref_loss = criterion(ref_output, ref_labels)
    ref_loss.backward()
    g_ref = parameters_to_vector([p.grad for p in params]).clone()

    # 3. Projection A-GEM
    dot_product = torch.dot(g, g_ref)
    if dot_product < 0:
        g_proj = g - (dot_product / torch.dot(g_ref, g_ref)) * g_ref
        g = g_proj

    # 4. Injection manuelle du gradient projeté dans le modèle
    optimizer.zero_grad()
    index = 0
    for p in params:
        num_params = p.numel()
        p.grad = g[index:index + num_params].view(p.shape).clone()
        index += num_params

    optimizer.step()
    return current_loss.item()
```

```python


def train_agrem_step(model,optimizer,criterion,current_batch,batch_ref,device):


  ## Initialisations

  current_imgs, current_labels=current_batch
  current_imgs, current_labels=current_imgs.to(device),current_labels.to(device)
  img_ref,label_ref=batch_ref
  img_ref,label_ref=img_ref.to(device),label_ref.to(device)

#calcul de g sur tache actuelle
  optimizer.zero_grad()
  current_output=model(current_imgs)
  current_loss=criterion(current_output, current_labels)
  current_loss.backward()

  #selection des parametres qui ont besoin de gradient
  param_with_grad1=[p.grad for p in model.parameters() if p.requires_grad ]


  # Transf en vecteur 1D
  g=parameters_to_vector(param_with_grad1).clone()


# calcul de g_ref
  optimizer.zero_grad()
  #img_ref,label_ref=batch_ref
  output_ref=model(img_ref)
  loss_ref=criterion(output_ref,label_ref)
  loss_ref.backward()

  param_with_grad2=[p.grad for p in model.parameters() if p.requires_grad ]
  g_ref=parameters_to_vector(param_with_grad2).clone()

  #produit scalire entre g et g_ref
  dot_product=torch.dot(g,g_ref)

  # Projection A-GEM
  if dot_product <0:
    terme_bas=torch.dot(g_ref,g_ref)
    coeff=dot_product/terme_bas

    # to calculate g - ( (g^t * g_ref) / (g_ref^T*g_ref) ) * g_ref
    g_proj=g - (coeff*g_ref )

    g=g_proj

#Use in the Resnet model
  # we get out of g_ref
  optimizer.zero_grad()

  index=0

  for p in model.parameters():

    # On regarde juste les couches qui ont besoin de gradient
    if p.requires_grad:

      # on compte le nombre d'elemnts de la couche ([3,3,4]=>3*3*4)
      num_params=p.numel()
      # on decoupe les tranches dans notre g
      tranche=g[index:index +num_params]

      p.grad=tranche.view(p.shape).clone()
      index+=num_params

  optimizer.step()

  return current_loss.item()







```

```python
def train_standard_step(model, optimizer,criterion,current_batch,device):
  """
  Effectue l'entrzinement de base (cad sans A-GEM)de 1 pas. On l'utilise suelemtn la toute premiere fois.
  """

  current_imgs,current_labels=current_batch
  # Remise a zero de sgradients
  optimizer.zero_grad()


  predictions=model(current_imgs)

  #comparer predictions et labels
  loss=criterion(predictions,current_labels)

  loss.backward()
  optimizer.step()

  return loss.item()
```

```python
def evaluate_continual_eval(model, test_loaders, list_of_learned_task, device):
    """
    Évalue la précision et la loss du modèle sur toutes les tâches apprises.
    Note : utilise l'objet 'criterion' défini globalement pour le calcul de la loss.
    """
    model.eval()
    results = {}  # Format: {id_task: {'acc': val, 'loss': val}}

    with torch.no_grad():
        for id_task in list_of_learned_task:
            specific_test_loader = test_loaders[id_task]
            total_correct = 0
            total_loss = 0.0
            total_samples = 0
            # recuperer les neurones valides pour la tache
            task_classes = tasks_definition[id_task]
            valid_neurons = [global_map[c] for c in task_classes]

            for images, labels in specific_test_loader:
                images, labels = images.to(device), labels.to(device)

                logits = model(images)
                # --- AJOUT : Masquage des neurones des autres tâches ---
                # On crée un masque rempli de -1e9 (moins l'infini)
                mask = torch.full_like(logits, -1e9)
                # On réactive uniquement les neurones de la tâche courante (valeur 0)
                mask[:, valid_neurons] = 0
                # On additionne : les neurones hors-tâche deviennent ultra-négatifs
                masked_logits = logits + mask

                loss = criterion(logits, labels)

                _, preds = torch.max(masked_logits, dim=1)
                total_correct += (preds == labels).sum().item()
                total_loss += loss.item() * labels.size(0)
                total_samples += labels.size(0)

            results[id_task] = {
                'acc': total_correct / total_samples,
                'loss': total_loss / total_samples
            }

    model.train()
    return results
```

```python
from tqdm import tqdm

def train_agem_global(model, optimizer, criterion, dict_test_loader, device='cuda'):
  buffer_agem = EpisodicMemory(max_memories_sample=memory_size_per_task)
  list_of_learned_task = []
  model.to(device)

  for id_task, classes in tasks_definition.items():
    print(f"\n--- Tâche {id_task} (Classes: {classes}) ---")

    current_df = label_ABCD_train[label_ABCD_train['label'].isin(classes)]
    current_dataset = IntelDataset(current_df, transform=my_transf, label_map=global_map)
    current_loader = DataLoader(current_dataset, batch_size=batch_size, shuffle=True)

    for epoch in range(1, epochs_per_task + 1):
      model.train()
      for imgs, labels in tqdm(current_loader, desc=f"Epoch {epoch}"):
        imgs, labels = imgs.to(device), labels.to(device)
        batch_ref = buffer_agem.get_reference_batch(batch_size=batch_size)

        if batch_ref is not None:
          train_agrem_step(model, optimizer, criterion, (imgs, labels), batch_ref, device)
        else:
          train_standard_step(model, optimizer, criterion, (imgs, labels), device)

    buffer_agem.store_memories(current_dataset)
    list_of_learned_task.append(id_task)

    # Évaluation avec la fonction dont la signature est préservée
    results = evaluate_continual_eval(model, dict_test_loader, list_of_learned_task, device)
    print(f"\nRésultats après Tâche {id_task} :")
    for t_id, metrics in results.items():
      print(f"   * Tâche {t_id} -> Précision: {metrics['acc']*100:.2f}% | Loss: {metrics['loss']:.4f}")
```
